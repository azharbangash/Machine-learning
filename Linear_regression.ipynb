```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/HP.csv")
df
```

```python
plt.xlabel("HP")
plt.ylabel("Price")
plt.scatter(df.area,df.prices)
```

```python
new_df=df.drop('prices',axis='columns')
new_df
```

```python
prices=df.prices
prices
```

```python
reg=linear_model.LinearRegression()
reg.fit(new_df,prices)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
